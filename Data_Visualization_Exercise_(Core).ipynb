**--> VISUALIZING ALL SKILLS RANKED BY ALL INCLUDED COUNTRIES WOULD BE INFORMATION OVERLOAD**

**--> DECIDED TO FOCUS ON UNITED STATES ONLY**

```python
#readable list of US-only data skills
skillset = df[(df['Country'] == 'United States')]['Skill'].unique()
for item in skillset:
  print(item)
```

```python
#see all US ranked skills
df[(df['Country'] == 'United States')]
```

**--> FOR 2016, THERE WERE ONLY 10 RANKED DATA SKILLS**

**--> TO STANDARDIZE THE VISUALIZATIONS, OPTED TO RESTRICT RANKED SKILLS TO THE TOP 10**

```python
#slim down to top 10 by year
#top 10 data skills for US for 2014
df2014 = df[(df['Country'] == 'United States') & (df['Year'] == 2014) & (df['Rank'] < 11)]
df2014 = df2014.sort_values('Skill')
df2014
```

```python
#top 10 data skills for US for 2015
df2015 = df[(df['Country'] == 'United States') & (df['Year'] == 2015) & (df['Rank'] < 11)]
df2015 = df2015.sort_values('Skill')
df2015
```

```python
#top 10 data skills for US for 2016
df2016 = df[(df['Country'] == 'United States') & (df['Year'] == 2016) & (df['Rank'] < 11)]
df2016 = df2016.sort_values('Skill')
df2016
```

**--> DECIDED TO CREATE 3 HORIZONTAL BAR CHARTS, ONE FOR EACH YEAR**

**--> SKILLS WOULD BE EASIER TO READ IF DISPLAYED ON THE Y-AXIS**

**--> PACKAGING THE BAR CHARTS AS THREE ROW / ONE COLUMN SUBPLOT WOULD ALLOW EASY COMPARISON BETWEEN YEARS**

```python
plt.figure(figsize=(20,12))
# bar plot with matplotlib

plt.subplot(3, 1, 1)
plt.barh('Skill', 'Rank', data=df2014,  label = '2014')
plt.xlabel('Rank', size=15)
plt.ylabel('Skill', size=15)
plt.title('United States Data Skills Top Ten 2014', size=18)


plt.subplot(3, 1, 2)
plt.barh('Skill', 'Rank', data=df2015,  label = '2015')
plt.xlabel('Rank', size=15)
plt.ylabel('Skill', size=15)
plt.title('United States Data Skills Top Ten 2015', size=18)


plt.subplot(3, 1, 3)
plt.barh('Skill', 'Rank', data=df2016,  label = '2016')
plt.xlabel('Rank', size=15)
plt.ylabel('Skill', size=15)
plt.title('United States Data Skills Top Ten 2016', size=18)

plt.tight_layout()
```
